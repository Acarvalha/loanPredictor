### (a) Introduction to the Dataset & Domain

- Nowadays, the **insurance domain** has been the most frequent usage of the **analytics & data science** methods in all industries.

- This dataset comes from a **real-world insurance domain**, which provides the data for **loan prediction**.
  It contains **615 records** and **13 features** describing personal and financial information of applicants who applied for a loan.

- The goal is to build a **classification model** that can automatically decide whether an applicant is **eligible for a loan** based on their **demographic and credit details**.

- **Automating** this process can help solve financial problems, make the process **faster and more consistent**, while also identifying **customer segments** more likely to be eligible.



### (b) Description of Features / Attributes

| Feature Name | Description |
|---------------|-------------|
| **Loan_ID** | The ID number of the loan for the applicant (unique identifier). |
| **Gender** | The gender of the applicant (**Male / Female**). |
| **Married** | The marital status of the applicant (**Yes / No**). |
| **Dependents** | Number of dependents the applicant has. |
| **Education** | The education level of the applicant (**Graduate / Not Graduate**). |
| **Self_Employed** | Whether the applicant is self-employed (**Yes / No**). |
| **ApplicantIncome** | Income of the applicant (numeric). |
| **CoapplicantIncome** | Income of the co-applicant (numeric). |
| **LoanAmount** | The amount of the loan applied for. |
| **Loan_Amount_Term** | The term (duration) of the loan in days. |
| **Credit_History** | The applicantâ€™s credit history (**1 = good, 0 = bad**). |
| **Property_Area** | The area type of the property (**Urban / Semi-urban / Rural**). |
| **Loan_Status** | Loan approval status (**Y = approved / N = rejected**). |
